## Problem 1 - Sum List
Write a function named `sum_list` that takes in a single parameter (a list) and returns the sum.  Once finished, you should call the function and print the value that is returned.  A list is provided for you to use as an argument. *Note*: The parameter for your function should be named something other than `my_list`.

```python
my_list = [x for x in range(1,10)]

def sum_list(list):
    sum = 0
    for x in list:
        sum += x
    return sum

sum_list(my_list)
```

## Problem 2
Write a function that finds the smallest of three numbers passed in as arguments. The function should print the smallest number.  Once you finish the function, call it three times with three different sets of numbers: one where the first argument is the smallest, one where the second is the smallest, and one where the third is the smallest. *You are not allowed to use the built-in min function.*

```python
def find_smallest(a,b,c):
    if a < b and a < c:
        print(a)
    elif b < a and b < c:
        print(b)
    else:
        print(c)

find_smallest(1,2,3)
find_smallest(2,1,3)
find_smallest(3,2,1)

```

## Problem 3
Write a function that takes a single string as a parameter.  This function should reverse the string and print it out.  The parameter should be a named parameter that has a default value of an empty string.  Once finished, you should call the function with at least two different words.


```python
def reverse_string(phrase):
    print(phrase[::-1])

reverse_string("book")
reverse_string("hello this is a string")
```

## Problem 4
Write a function that takes a single string as a parameter.  The function should then total up the number of lowercase and uppercase characters.  It should also print these totals in a clear way.  Be sure to call the function with a couple of sample words once you finish.

```python
def add_up(phrase):
    upper = 0
    lower = 0
    for i in range(0,len(phrase)):
        if phrase[i].isupper():
            upper += 1
        else:
            lower += 1
    print(upper)
    print(lower)
    print()
add_up("BaNaNa")
add_up("BANANA")
add_up("banana")
        
```

## Problem 5
Write a function that takes a string and determines if it is a pangram or not and prints the result.  A string is a pangram if it contains all of the letters of the alphabet.

```python
def is_pangram(phrase):
    all_letters = "qwertyuiopasdfghjklzxcvbnm"
    string = phrase.lower()
    
    for i in range(0,len(all_letters)):
        valid = False
        for j in range(0, len(string)):
            if string[j] == all_letters[i]:            
                valid = True
        if not valid:
            return False
    return True 
print(is_pangram("Sphinx of black quartz, judge my vow."))
print(is_pangram("banana"))
```
