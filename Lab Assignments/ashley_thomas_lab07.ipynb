## Problem 7 - Another Odd List
Create a new list that contains all of the odd numbers between 1 and 20, each divided by 2.  You may not hard-code the values and must use a __for loop__. After, use the __print__ function to output your list.

```python
x = []
for i in range(20):
    if i % 2 != 0:
        x.append(i/2)
for i in x:
    print(i)
```

## Problem 8 - List Comprehension
Create a new list that contains all of the odd numbers between 1 and 20, each divided by 2.  You may not hard-code the values and must use __list comprehension__. After, use the __print__ function to output your list.

```python
x = [i/2 for i in range(20) if i % 2 != 0 ]
for i in x:
    print(i)
```
