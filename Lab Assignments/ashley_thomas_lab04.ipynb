## Problem 2
Write a while loop to print all numbers between 50 and 5000 that is a multiple of 999.

```python
x = 50
while x <= 5000:
    if x % 999 == 0:
        print(x)
    x += 1
```

## Problem 3
Below is the guessing game from the notes.  Modify it to limit the number of guesses the user can make to __3__.  If the user takes up all 3 guesses, then tell them "You have no more guesses left. You lose!" and then stop the game.

```python
import random

num = random.randint(0,50)
guess = -1
num_guess = 1


print(f"num = {num}")
while guess != num & num_guess <= 3 :    
    guess = int(input("Please guess a number between 0 and 50: "))
    num_guess += 1
    if guess > 100 or guess < 0:
        print("Invalid number.")
    elif guess > num:
        print("Your number is too high!")
    elif guess < num:
        print("Your number is too low!")
    else:
        print("That is correct! You win!")
    


```
